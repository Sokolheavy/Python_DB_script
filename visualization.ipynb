```python
import pandas as pd
import numpy as np

import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 16, 8
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nsethi31/German-Credit-Data/master/GermanCredit.csv')
```

```python
df.columns
```

```python
for i in df.columns:
    print('\nFrequency count for variable %s'%i)
    print(df[i].value_counts())
```

## Histogram :

```python
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(df['AGE'],bins = 7) # Here you can play with number of bins Labels and Tit
plt.title('Age distribution')
plt.xlabel('AGE')
plt.ylabel('#Employee')
plt.show()
```

## Box Plot:

```python
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(df['AGE'])
plt.show()
```

## Violin Plot:

```python
sns.violinplot(df['NUM_CREDITS'], df['AGE']) #Variable Plot
sns.despine()
```

## Bar Chart:

```python
var = df.groupby('NUM_CREDITS').AMOUNT.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('NUM_CREDITS')
ax1.set_ylabel('Sum of Sales')
ax1.set_title("NUM_CREDITS wise Sum of AMOUNT")
var.plot(kind='bar')
```

## Line Chart:

```python
var = df.groupby('NUM_CREDITS').AMOUNT.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('NUM_CREDITS')
ax1.set_ylabel('Sum of AMOUNT')
ax1.set_title("NUM_CREDITS wise Sum of AMOUNT")
var.plot(kind='line')
```

## Stacked Column Chart

```python
var = df.groupby(['NUM_CREDITS','RESPONSE']).AMOUNT.sum()
var.unstack().plot(kind='bar',stacked=True,  color=['red','blue'], grid=False)
```

## Scatter Plot:

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['DURATION'],df['AMOUNT']) 
plt.show()
```

## Bubble Plot:

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['AGE'],df['NUM_CREDITS'], s=df['AMOUNT'])
plt.show()
```

## Heat Map:

```python
#correlation matrix
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```
