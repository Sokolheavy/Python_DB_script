```pyspark3
bin_col = [i for i in data.columns if len(data[i].unique())==2]

le = preprocessing.LabelEncoder()
for i in bin_col:
    data[i] = le.fit_transform(data[i])
    
int_col = ['MonthlyCharges', 'TotalCharges', 'tenure']
for i in int_col:
    data[i] = data[i].astype(float)

one_hot = [i for i in data.columns if (i not in bin_col) and (i not in int_col)]
one_hot_data = pd.get_dummies(data[one_hot])  
data_final = pd.concat([data, one_hot_data], axis = 1)
data_final.drop(one_hot, axis = 1, inplace = True)
```

```pyspark3
X_train, X_test, y_train, y_test = train_test_split(data_final.loc[:, data_final.columns!='Churn'], data_final['Churn'] , test_size=0.33, random_state=42)
```
