```python
period = 26
previous_df = pd.DataFrame()

for i in SKU_Customer_list:
    model = auto_arima(data[data.SKU_Customer == i].Sales, 
                         start_p = 1,
                         start_q = 1,
                         max_p = 3,
                         max_q = 3,
                         m=12,
                         start_P = 0,
                         seasonal = True,
                         d =1,
                         D = 1,
                         trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)
    
    model.fit(data[data.SKU_Customer == i].Sales)

    future_forecast, confint  = model.predict(n_periods = period, return_conf_int=True)
    last_df = pd.DataFrame({'SKU_Customer': i*period,
                            'prediction': future_forecast,
                            'conf_low': confint[ :,0],
                            'conf_upper': confint[ :,1]})
    
    previous_df = pd.concat([previous_df, last_df], ignore_index = True)  
```

```python
y_pred = previous_df.sort_values(by=['SKU_Customer'])
```

```python
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs(y_true - y_pred)) * 100 / np.mean(np.abs(y_true))
```

```python
def mape_by_month_LE(y_true, y_pred): 
    test2 = test.copy()
    test2['Sales_pred'] = y_pred
    test2['Sales_pred'] = test2.groupby(['Year', 'Month_No', 'SKU_Customer']).Sales_pred.transform('mean')
    return mape(test['Sales'], test2['Sales_pred'])
```
